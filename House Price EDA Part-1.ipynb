```python
#looking missing values
dataset.isnull().sum()
```

```python
#looked the features i.e columns and giving conditions that if dataset having more than 1 value has missing values
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>=1]
```

```python
#these are features with nan values
features_with_nan
```

```python
#dataset['Electrical'].isnull().mean()
for feature in features_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values')
    
```

```python
sns.boxplot(dataset['SalePrice'])
```

```python
#plot amazing diagrams
data=dataset.copy()
for feature in features_with_nan:
    
    #lets make a variable that indicate 1 if missing and 0 otherwise
    data[feature]=np.where(data[feature].isnull(),1,0)
    #lets calculate the median saleprice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and dependent variable is clearly visible.So we need to replace these nan values with some meaningful which we will do in the Feature Engineering Section

# Numerical variables


```python
dataset.info()
```

```python
#List of numerical variables
dataset['SaleType'].dtypes!='object'
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype!="object"]
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
dataset.info()
```

```python
## Temporal Variables(Eg datetime variables)
year_feature=[feature for feature in numerical_features if 'Yr' in feature or "Year" in feature]
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs Year SOld")
```

```python
###Here we will compare the difference between all years features with salesprice
data=dataset.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]#age of house
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("salePrice")
        plt.show()
```

With all the yearwise information,it is clearly visible tha the new house are more costlier than the old one followinf power lae or pareto distribution

```python
##numerical variable-2 types
#continous and discrete selected as 25 less is discrete

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<=25]
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
##lets find the relationship between discrete and sales price
data=dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    #plt.scatter(data[feature],data["SalePrice"])
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

##There is relationship between Discrete feature and salesPrice

```python
##Continuous Variable
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature]
print(len(continuous_feature))
```

```python
## Lets analyze the continous values by creating histogram to understand the distribution
data=dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)["SalePrice"].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

```python
data=dataset.copy()
for feature in continuous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Continuous")
    plt.title(feature)
    plt.show()
```

# Eda Part 2

```python
## we will be using logarathmic transformation
data=dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature)
        plt.show()
```

```python
##outliers
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

##Categorical Variables

```python
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
print(len(categorical_feature))
```

```python
dataset[categorical_feature].head()
```

```python
for feature in categorical_feature:
    print("The feature name is {} and the number of categories are {}".format(feature,len(dataset[feature].unique())))
```

```python
##find the relationship between categorical feature and saleprice
data=dataset.copy()
for feature in categorical_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

There is a relationship between categorical value and saleprice
